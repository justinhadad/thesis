```python
import numpy as np
import math
import random
from copy import copy, deepcopy
import pandas as pd  
from matplotlib import pyplot as plt


def runauction(x):
    n = x
    s = 10

    employees = [[random.uniform(0,2) for i in range(s)] for j in range(n)]
    left=n
    wage=0
    employee_number=100
    occupied = []
    cost=0
    
    wages=[]
    costs=[]

    for j in range(s):
            minimum=100
            for i in range(n):
                if employees[i][j]<minimum and i not in occupied:
                    minimum=employees[i][j]
                    employee_number=i
            wages.append(2-((left-1)/left)*(2-minimum))
            costs.append(minimum)
            left-=1
            occupied.append(employee_number)
    _sum=0
    
    for i in range(len(wages)):
        _sum+=wages[i]
    average_wage=_sum/s
    
    for i in range(len(costs)):
        _sum+=costs[i]
    average_cost=_sum/s
    
    return(average_wage, average_cost)

def main():
    
#     print(sum/n)


    
    iterations=100
    limit_emp = 100
    avg_cost=[0]*limit_emp
    avg_wage=[0]*limit_emp
    
    for _iter in range(iterations):
        for _x in range(10,limit_emp,1):
            average_wage, average_cost = runauction(_x)
            avg_cost[_x]+=average_cost
            avg_wage[_x]+=average_wage
    
    for item in range(len(avg_cost)):
        avg_cost[item]=avg_cost[item]/iterations
        avg_wage[item]=avg_wage[item]/iterations

    absenteeism=[]
    for cost in range(len(avg_cost)):
        absenteeism.append(0.06*avg_cost[cost]/(avg_wage[cost]+0.00000000001))
    
    plt.plot(absenteeism)
    plt.title('Absenteeism in Dutch trades \n S=10, N=[10,100], 100 iterations')
    plt.xlabel('Number of Employees')
    plt.ylabel('Absenteeism Rate')
    plt.xlim([10, limit_emp])
    plt.show()
    print
    

    
#     plt.plot(avg_cost)
#     plt.title('Average cost of each assignment \n S=10, N=[10,100], 100 iterations')
#     plt.xlabel('Number of Employees')
#     plt.ylabel('Cost')
#     plt.xlim([10, limit_emp])
#     plt.show()
        
#     for _iter in range(iterations):
#         for _x in range(11,limit_emp,1):
#             avg_wages[_x]+=(runauction(_x))
    
    
    
#     for _iter in range(iterations):
#         for _x in range(10,limit_emp,1):
#             avg_wages[_x]+=(runauction(_x))
        
#     for item in range(len(avg_wages)):
#         avg_wages[item]=avg_wages[item]/iterations

    
main()
```
