```python
import matplotlib.pyplot as plt
import numpy as np

v=0
x=np.arange(2, 10, 1)
y1= (1/2)**x
y2= (3/4)**((x*(x-1))/2+(x+v-x)*x)
plt.plot(x, y1, label='Acceptability', color="black")
plt.plot(x, y2, label='Stability')
plt.legend()


plt.title('Random Assignment, N=S')
plt.xlabel('Shifts')
plt.ylabel('Probability')
plt.show()

v=1
y2= (3/4)**((x*(x-1))/2+(x+v-x)*x)

plt.plot(x, y1, label='Acceptability', color="black")
plt.plot(x, y2, label='Stability, N=S+1')
plt.legend()

plt.title('Random Assignment, N>S')
plt.xlabel('Shifts')
plt.ylabel('Probability')

v=2
y2= (3/4)**((x*(x-1))/2+(x+v-x)*x)
plt.plot(x, y2, label='Stability, N=S+2')
plt.legend()

v=3
y2= (3/4)**((x*(x-1))/2+(x+v-x)*x)
plt.plot(x, y2, label='Stability, N=S+3')
plt.legend()

v=4
y2= (3/4)**((x*(x-1))/2+(x+v-x)*x)
plt.plot(x, y2, label='Stability, N=S+4')
plt.legend()

v=5
y2= (3/4)**((x*(x-1))/2+(x+v-x)*x)
plt.plot(x, y2, label='Stability, N=S+5')
plt.legend()

v=10
y2= (3/4)**((x*(x-1))/2+(x+v-x)*x)
plt.plot(x, y2, label='Stability, N=S+10')
plt.legend()


plt.show()
```
