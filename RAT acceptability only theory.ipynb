```python
import matplotlib.pyplot as plt
import numpy as np
import math


def myFun(x):
    value=0
    total=0
    for i in range(1,x):
        value=math.factorial(x)/(math.factorial(i)*math.factorial(x-i))*(1/2)**x*(1-(x-i)*5/16)**i
        total+=value
    return total+1/2**x
        

f2 = np.vectorize(myFun)
x=np.arange(2, 10, 1)

y1= (1/2)**x
                    
plt.plot(x, y1, label='Acceptability before trades', color="black")
plt.plot(x, f2(x), label='Acceptability after trades')
plt.legend()

plt.xlabel('Shifts')
plt.ylabel('Probability')
plt.ylim([0, 1])

plt.show()
```
